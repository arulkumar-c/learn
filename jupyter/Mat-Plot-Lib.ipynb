```python

months = ['Jan', 'Feb', 'March', 'Apr', 'May', 'June']
sales = [10789, 12897, 13554, 12650, 14320, 15236]

plt.plot(months,sales,linewidth=3)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales');
```

```python
salesA = [10789, 12897, 13554, 13650, 14320, 15236]
salesB = [10364, 13326, 14569, 14987, 15005, 15241]
salesC = [9891, 11369, 11659, 12458, 12430, 13985]

plt.plot(months,salesA,linewidth=3)
plt.plot(months,salesB,linewidth=1)
plt.plot(months,salesC,linewidth=1)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.legend(labels=['Company A', 'Company B', 'Company C'], loc='lower right');
```

```python
plt.plot(months,salesA,linewidth=2)
plt.plot(months,salesB,linewidth=2)
plt.plot(months,salesC,linewidth=2)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.legend(labels=['Company A', 'Company B', 'Company C'], loc='upper left')
plt.grid(True);
```

```python
plt.plot(months,salesA,linewidth=2,marker='o')
plt.plot(months,salesB,linewidth=2,marker='o')
plt.plot(months,salesC,linewidth=2,marker='o')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.legend(labels=['Company A', 'Company B', 'Company C'], loc='upper left')
plt.grid(color='red', alpha=0.2, linewidth=2);
```

```python
x = np.linspace(0, 2, 200)

fig = plt.figure(figsize=(12, 6))

plt.plot(x, x, label='linear',linewidth=1)
plt.plot(x, x**2, label='quadratic',linewidth=1)
plt.plot(x, x**3, label='cubic',linewidth=1)

plt.grid(True)
plt.title("Simple Plot")
plt.legend();
```

```python
!pwd
!ls -l ./train.csv

df = pd.read_csv('./train.csv')
df.head()
```

```python
age_m = df[df['Sex'] == 'male']['Age']
fare_m = df[df['Sex'] == 'male']['Fare']
age_f = df[df['Sex'] == 'female']['Age']
fare_f = df[df['Sex'] == 'female']['Fare']

fig = plt.figure(figsize=(12, 6))
plt.scatter(x=age_m, y=fare_m,s=25)
plt.scatter(x=age_f, y=fare_f,s=25)
plt.legend(['male', 'female']);
```

```python
pclass = df[df['Sex']=='male']['Pclass']

fig = plt.figure(figsize=(12, 6))
plt.scatter(x=age_m, y=fare_m,s=25, c=pclass, cmap='cool')
plt.colorbar(ticks=[1, 2, 3])

plt.xlabel('Age - Male')
plt.ylabel('Fare');
```

```python
emb = df['Embarked'].value_counts()

plt.bar(x=emb.index, height=emb.values, width=0.5);
```

```python
survivor = df['Survived'].value_counts()

plt.bar(x=[0,1], height=survivor.values, width=0.2, color='yellow', edgecolor='red',linewidth=2);
print('Thus the majority didnt survive.')
```

```python
m_survivors = df[df['Sex']=='male']['Survived'].value_counts()
f_survivors = df[df['Sex']=='female']['Survived'].value_counts()

fig, axes = plt.subplots(1, 2) #for subplots
fig.set_size_inches(8,4) 

axes[0].barh(y=m_survivors.index, width=m_survivors.values, height = 0.2, color=['orange', 'blue'])
axes[0].set_title('Male Survivors')


axes[1].barh(y=f_survivors.index, width=f_survivors.values, height = 0.2, color=['blue', 'orange'])
axes[1].set_title('Female Survivors');
```

```python
embarked_m = df[df['Sex']=='male']['Embarked'].value_counts()
embarked_f = df[df['Sex']=='female']['Embarked'].value_counts()

plt.bar(x=embarked_m.index, height=embarked_m.values)
plt.bar(x=embarked_f.index, height=embarked_f.values, bottom=embarked_m.values, color='orange')

plt.xlabel('Port')
plt.ylabel('Passengers')
plt.title('Embarking Ports by Sex')
plt.legend(labels=['Male', 'Female']);
```

```python
X = np.linspace(start=-np.pi, stop=np.pi, num=200)
Y = np.sin(2*X)

fig = plt.figure(figsize=(8, 5))
plt.plot(X,Y)
plt.text(-1.80, 0.5, 'y = sin(2x)', fontsize=14);
```

# Fibonacci Series - Regular and Cached Approach

## 1. Regular Approach

```python
def fib(n: int) -> int:
    if n <= 1:
        return(1)
    return(fib(n - 2) + fib(n - 1))

#print(f'fib(100) = {fib(50)}')
#print(f'fib(50) = {fib(50)}')
print(f'fib(10) = {fib(10)}')
```

## 2. Cached Approach

```python
fib_cache = {}

def fib(n: int) -> int:
    global fib_cache

    if n not in fib_cache:
        if n <= 1:
            fib_cache |= {n: 1}
        else:
            fib_cache |= {n: fib(n - 2) + fib(n - 1)}
        print(f'fib({n}) = {fib_cache[n]}')

    return(fib_cache[n])

print(f'fib(2000) = {fib(2000)}')
```

## Power of 2 - Using bit shift and using pow()
